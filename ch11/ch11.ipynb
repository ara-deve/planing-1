## 1. 성과 측정의 본질

피터 드러커: *"측정되지 않으면 관리되지 않는다."*

기획서에 KPI가 없으면 실행 단계에서 진행 상황을 알 수 없다. 그러나 **잘못 측정하면 잘못된 행동을 부른다**. 측정 지표가 행동을 결정하기 때문이다.

사례:

- 콜센터에서 "통화 시간"만 KPI로 잡으면 직원들은 빨리 끊기 시작한다 → 고객 만족도 하락.
- 학교에서 "평균 점수"만 KPI로 잡으면 어려운 학생을 포기하기 쉽다 → 형평성 악화.

**기획자의 질문**

> 우리 KPI가 의도하지 않은 행동을 유발하지 않는가?

## 2. 선행 지표 vs 후행 지표

| 구분 | 선행 지표 (Leading) | 후행 지표 (Lagging) |
|------|-------------------|--------------------|
| 시점 | 결과 전 | 결과 후 |
| 통제 가능성 | 통제 가능 | 통제 불가 |
| 다이어트 | 일일 칼로리 섭취 | 체중 |
| 매출 | 영업 미팅 수 | 분기 매출 |
| 정책 | 시범 신청 건수 | 1년 후 고용률 |

한 줄 요약:

> 후행 지표만 보면 손쓸 시점에는 이미 늦는다. **선행 지표를 함께 두자**.

균형 비율 권장: 선행 50% + 후행 50%.

## 3. SMART 매뉴얼

KPI를 작성할 때 다음 5가지를 점검한다.

| 글자 | 점검 질문 |
|------|---------|
| **S**pecific | 무엇을 측정하는지 한 문장으로 적을 수 있는가 |
| **M**easurable | 숫자로 측정 가능한가 |
| **A**chievable | 달성 가능한 수준인가 |
| **R**elevant | 핵심 목표와 연결되는가 |
| **T**ime-bound | 언제까지인가 |

**나쁜 KPI** → "고객 만족도 향상"

**좋은 KPI** → "2026년 12월까지 NPS 30점 달성 (현재 18점)"

## 5. 이상 탐지의 직관

복잡한 알고리즘 없이도 두 가지 방법으로 충분히 잡을 수 있다.

1. **이동 평균 (Moving Average)**: 최근 N주 평균. 단기 흐름이 추세에서 어긋나는지 본다.
2. **표준편차 밴드 (±2σ)**: 평균에서 표준편차 2배를 벗어나면 이상 의심.

```
Alert  if  현재 값  >  이동평균 + 2 * 이동표준편차
       or  현재 값  <  이동평균 - 2 * 이동표준편차
Watch  if  현재 값  >  이동평균 + 1 * 이동표준편차
Good   otherwise
```

권장 윈도우: 주간 데이터는 8주, 월간 데이터는 6개월.

**기획자의 주의사항**

- Alert가 자주 뜨면 임계값(σ 배수)을 키운다.
- 계절성(분기별·요일별 패턴)이 있으면 이동 평균만으로는 부족하다 → 같은 분기끼리 비교.

## 6. KPI 대시보드 매뉴얼

**대시보드 작성 6단계**

1. 핵심 KPI **5~7개**를 고른다 (너무 많으면 안 본다).
2. 각 KPI에 **목표값**과 **임계값**을 설정한다.
3. 색상 대신 **텍스트로 상태**를 표시한다 (Good / Watch / Alert) — 색맹·인쇄·요약문 호환을 위해.
4. 한 화면에 다 보이도록 한다.
5. 임원 보고용은 후행 + 선행 1개씩, 실무용은 선행 위주.
6. 매주 같은 시간에 갱신·검토한다.

## 7. 바이브 코딩 프롬프트

````python
# 1단계 코드: 데이터 탐색
import pandas as pd
import matplotlib.pyplot as plt

df_app = pd.read_csv("data/weekly_applications.csv")

print("=" * 70)
print("1단계: 데이터 탐색 — 주별 파일럿 신청 건수")
print("=" * 70)
print("\n📊 데이터 요약:")
print(f"  • 총 주차: {len(df_app)}주")
print(f"  • 컬럼: {df_app.columns.tolist()}")

print("\n📈 신청 건수 (applications) 기초 통계:")
print(f"  • 평균: {df_app['applications'].mean():.1f}건/주")
max_idx = df_app['applications'].idxmax()
min_idx = df_app['applications'].idxmin()
print(f"  • 최대: {df_app['applications'].max():.0f}건 (주차: {int(df_app.loc[max_idx, 'week'])})")
print(f"  • 최소: {df_app['applications'].min():.0f}건 (주차: {int(df_app.loc[min_idx, 'week'])})")
print(f"  • 표준편차: {df_app['applications'].std():.1f}건")

print("\n⚠️  이상 탐지 결과 (Status):")
print(f"  • Good:  {(df_app['status'] == 'Good').sum()}주")
print(f"  • Watch: {(df_app['status'] == 'Watch').sum()}주")
print(f"  • Alert: {(df_app['status'] == 'Alert').sum()}주")

print("\n🚨 Alert 주차 목록:")
alert_weeks = df_app[df_app['status'] == 'Alert']['week'].tolist()
if len(alert_weeks) > 0:
    print(f"  → {alert_weeks}")
else:
    print("  → 없음")

print("\n📋 데이터 샘플:")
print(df_app[["week", "applications", "ma8", "upper", "lower", "status"]].head())
````

````python
# 2단계 코드: 이상 탐지 시각화
import pandas as pd
import matplotlib.pyplot as plt

df_app = pd.read_csv("data/weekly_applications.csv")

plt.figure(figsize=(11, 5))
plt.plot(df_app["week"], df_app["applications"], color="black", linewidth=1.2, label="신청 건수", marker="o", markersize=3)
plt.plot(df_app["week"], df_app["ma8"], color="gray", linestyle="--", linewidth=1, label="MA(8)")
plt.fill_between(df_app["week"], df_app["lower"], df_app["upper"], color="lightgray", alpha=0.3, label="±2σ 밴드")

# Alert 포인트 강조
alerts = df_app[df_app["status"] == "Alert"]
plt.scatter(alerts["week"], alerts["applications"], s=100, facecolors="white", edgecolors="black", linewidths=1.5, label="Alert", zorder=5)

plt.xlabel("주차 (Week)", color="black", fontsize=11)
plt.ylabel("신청 건수 (명)", color="black", fontsize=11)
plt.title("파일럿 신청 건수 — 이동평균 + ±2σ 이상 탐지\n(청년 일자리 디지털 전환 지원 사업)", color="black", fontsize=12, fontweight="bold")
plt.legend(loc="upper left", framealpha=0.95)
plt.grid(True, alpha=0.2, linestyle=":")
plt.tight_layout()
plt.show()

print("✅ 2단계 완료: 이상 탐지 시각화 그래프 생성")
````

````python
# 3단계 코드: Alert 주차 찾기 + 원인 가설
import pandas as pd

df_app = pd.read_csv("data/weekly_applications.csv")

print("=" * 70)
print("3단계: 이상 시점 해석 — Alert 주차 분석 + 원인 가설")
print("=" * 70)

# Alert 주차 필터링
alert_data = df_app[df_app['status'] == 'Alert']

print("\n🚨 Alert 주차 상세 정보:")
if len(alert_data) > 0:
    print(alert_data[["week", "applications", "ma8", "upper", "lower"]].to_string(index=False))
    
    alert_weeks = alert_data['week'].tolist()
    print(f"\n📌 Alert 주차: {alert_weeks}")
    
    for week in alert_weeks:
        row = df_app[df_app['week'] == week].iloc[0]
        print(f"\n  주차 {int(week)}:")
        print(f"    - 신청 건수: {int(row['applications'])}명")
        print(f"    - 이동평균: {row['ma8']:.1f}명")
        print(f"    - 상한선(+2σ): {row['upper']:.1f}명")
        print(f"    - 하한선(-2σ): {row['lower']:.1f}명")
        
        if row['applications'] > row['upper']:
            print(f"    ⬆️  이상 유형: 급증 (평균 대비 {(row['applications'] - row['ma8']):.0f}명 초과)")
        else:
            print(f"    ⬇️  이상 유형: 급감 (평균 대비 {(row['ma8'] - row['applications']):.0f}명 미달)")

    print("\n💡 원인 가설 (청년 일자리 정책 맥락):")
    print("  1. 언론·SNS 홍보 효과: 대규모 매체 홍보, 인플루언서 협력 등으로 인한 수요 폭증")
    print("  2. 외부 정책 변수: 타 정부부처 청년지원금 종료, 대학 캠프 모집 시즌, 계절적 요인 등")
else:
    print("  → Alert 주차가 없습니다.")
````

````python
# 4단계 코드: 대시보드 해석 — Watch/Alert KPI 필터링 + 개선 방향 제안
import pandas as pd

dashboard = pd.read_csv("data/kpi_dashboard_snapshot.csv")

print("=" * 70)
print("4단계: 대시보드 해석 — Watch/Alert KPI 필터링 + 개선 방향")
print("=" * 70)

print("\n📊 전체 KPI 현황:")
print(dashboard.to_string(index=False))

# Watch 또는 Alert 상태 필터링
problem_kpis = dashboard[dashboard['status'].isin(['Watch', 'Alert'])]

print("\n\n⚠️  개선이 필요한 KPI (Watch/Alert):")
if len(problem_kpis) > 0:
    print(problem_kpis[["kpi", "current", "target", "status"]].to_string(index=False))
    
    print("\n💡 개선 방향 (한 문장 제안):")
    for idx, row in problem_kpis.iterrows():
        kpi_name = row['kpi']
        current = row['current']
        target = row['target']
        gap = target - current
        status = row['status']
        
        if gap > 0:
            print(f"  • {kpi_name} (현재 {current:.1f}, 목표 {target:.1f}):")
            if status == "Alert":
                print(f"    → 【긴급】 온보딩·교육 강화, 담당자 인센티브 제도 도입으로 {gap:.1f}p 개선 필요")
            else:
                print(f"    → 【주의】 프로세스 개선 및 개인별 멘토링으로 {gap:.1f}p 향상 추진")
        else:
            print(f"  • {kpi_name} (현재 {current:.1f}, 목표 {target:.1f}):")
            print(f"    → 【주의】 소비 절감 방안 검토 및 예산 효율화 전략 수립")
else:
    print("  ✅ Watch/Alert 상태인 KPI가 없습니다. 모든 KPI가 정상입니다!")

print("\n\n✅ 4단계 완료: 대시보드 해석 및 개선 방향 도출")
````
