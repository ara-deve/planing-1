## 과제 3. 간단한 시뮬레이션 (6점)

현재 상태(구독자 3,200명, 해지율 28%, 신규 120명/월)로 **6개월 후 구독자 수**를 계산하세요.

### 요구사항

1. **(3점)** 아래 두 시나리오의 6개월간 구독자 수 변화를 계산하세요:
   - 시나리오 A: 현상 유지 (해지율 28%)
   - 시나리오 B: 배송 개선으로 해지율 20%로 감소

2. **(3점)** 두 시나리오의 구독자 수 추이를 **라인 차트**로 비교하고, 6개월 후 차이를 `print()`로 출력하세요.

````python
# ✏️ 과제 3 풀이
# 시나리오 A vs B 시뮬레이션 + 라인 차트 + 결론 print()

# 초기값
initial_subscribers = 3200
new_users = 120
months = 6

# 시나리오 A (해지율 28%)
churn_A = 0.28

# 시나리오 B (해지율 20%)
churn_B = 0.20

subs_A = [initial_subscribers]
subs_B = [initial_subscribers]

# 계산
for i in range(months):
    next_A = subs_A[-1] * (1 - churn_A) + new_users
    next_B = subs_B[-1] * (1 - churn_B) + new_users
    
    subs_A.append(next_A)
    subs_B.append(next_B)

# x축 (개월)
x = list(range(0, months+1))

# 그래프
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x,
    y=subs_A,
    mode='lines+markers',
    name='시나리오 A',
    line=dict(color='black'),
    marker=dict(color='white', line=dict(color='black', width=2))
))

fig.add_trace(go.Scatter(
    x=x,
    y=subs_B,
    mode='lines+markers',
    name='시나리오 B',
    line=dict(color='gray'),
    marker=dict(color='white', line=dict(color='black', width=2))
))

fig.update_layout(
    title='구독자 수 변화 (6개월)',
    xaxis_title='개월',
    yaxis_title='구독자 수 (명)',
    font=dict(color='black')
)

fig.show()

# 6개월 후 차이 출력
difference = subs_B[-1] - subs_A[-1]
print(f"시나리오 A (해지율 28%) - 6개월 후 구독자 수: {int(subs_A[-1])}명")
print(f"시나리오 B (해지율 20%) - 6개월 후 구독자 수: {int(subs_B[-1])}명")
print(f"차이: {int(difference)}명")
````

### ✏️ 과제 3 해석 (이 셀에 직접 타이핑)

- 시나리오 A와 B의 6개월 후 구독자 수 차이는 몇 명인가?
- 이 차이가 **매출(월 구독료 5만 원)**로 환산하면 얼마인가?
- 이 결과가 경영진에게 시사하는 바를 1문장으로 작성하세요.

그래프를 보면 해지율이 높은 시나리오 A는 시간이 지날수록 구독자 수가 빠르게 줄어드는 것을 확인할 수 있다. 반면 시나리오 B는 감소 폭이 비교적 작아서 더 많은 구독자를 유지하는 모습을 보인다. 6개월 후 기준으로 약 466명 정도 차이가 발생하고, 매출로 보면 약 2,330만 원 정도의 차이가 나는 것으로 나타난다. 따라서 해지율을 낮추는 것이 고객 유지뿐만 아니라 매출에도 중요한 영향을 준다고 생각한다.
