## 1. 점 추정 vs 확률 분포

기획자에게 매출을 물으면 흔히 **하나의 숫자**로 답한다.

> 내년 매출은 1억입니다.

이 답은 깔끔하지만 위험하다. **불확실성이 사라진다**. 만약 실제 매출이 7천만이 나오면 "틀린 예측"이 된다.

정직한 답은 **분포**로 답하는 것이다.

> 내년 매출은 8천만~1.2억 사이일 가능성이 높습니다(80% 신뢰). 평균은 1억 정도입니다.

베이지안은 이 분포를 그대로 다룬다. 하나의 값이 아니라 **가능한 값들의 확률**을 다룬다.

| 사고방식 | 표현 | 장점 | 단점 |
|---------|------|------|------|
| 점 추정 | "매출 1억" | 단순, 결정 빠름 | 불확실성 무시 |
| 확률 분포 | "8천만~1.2억(80%)" | 위험 인식 | 약간의 학습 필요 |

## 2. 사전·가능도·사후, 한 문장으로

베이즈 정리는 한 줄이다. 수식은 외우지 않는다. 의미만 기억하면 된다.

$$P(H|D) = \frac{P(D|H) \times P(H)}{P(D)}$$

| 기호 | 이름 | 의미 |
|---|---|---|
| $P(H)$ | 사전확률 (Prior) | 데이터를 보기 전, 가설 H가 맞을 확률 |
| $P(D \mid H)$ | 가능도 (Likelihood) | 가설 H가 맞다면, 데이터 D가 관측될 확률. 의료 예시에서는 실제 환자가 양성 판정을 받을 확률 = 민감도(재현율) = TP/(TP+FN) |
| $P(D)$ | 전체확률 (Evidence) | 데이터 D가 관측될 전체 확률. = P(D\|H)×P(H) + P(D\|¬H)×P(¬H). 가설 H와 무관한 상수 |
| $P(H \mid D)$ | 사후확률 (Posterior) | 데이터 D를 본 후, 가설 H가 맞을 확률 |

한 문장으로 요약하면: **사후확률은 가능도와 사전확률의 곱에 비례한다.**

$$P(H|D) \propto P(D|H) \times P(H)$$

$P(D)$는 가설 H와 무관한 상수이므로, 사후확률의 상대적 크기는 가능도 × 사전확률로 결정된다.

**사례: 2022년 러시아의 우크라이나 침공**

사전·가능도·사후가 실제로 어떻게 작동하는지 보여주는 실제 사례다. 특히 **가능도를 무시했을 때 어떤 오류가 생기는지** 주목한다.

**배경**

2021년 말, 러시아가 우크라이나 국경 주변에 약 15만 명의 병력을 집결시켰다. 많은 서방 분석가들은 "설마 실제로 침공하겠어? 블러핑일 것"이라고 봤다. 2022년 2월, 러시아는 실제로 침공했다.

왜 많은 분석가들이 틀렸는가? 베이지안으로 보면 이유가 명확해진다.

**가설과 증거**

- **H (가설)**: 러시아가 우크라이나를 전면 침공한다
- **D (증거)**: 국경에 15만 병력이 집결됐다

| 개념 | 질문 | 이 사례 |
|---|---|---|
| 사전확률 P(H) | 증거를 보기 전, 침공 가능성은? | 냉전 이후 유럽에서 전면전은 없었다 → 분석가들이 **낮게** 봄 (~20%) |
| **가능도 P(D\|H)** | **침공한다면, 15만 병력 집결이 자연스러운가?** | **전면 침공에 15만 집결은 필수다 → 매우 높음 (~90%)** |
| 가능도 P(D\|¬H) | 침공 안 한다면, 15만 집결이 자연스러운가? | 블러핑치고는 규모가 지나치게 크다 → 낮음 (~15%) |
| 사후확률 P(H\|D) | 증거를 본 뒤, 침공 가능성은? | 베이즈로 계산하면 크게 올라가야 했다 (~60%) |

**가능도가 핵심이었던 이유**

많은 분석가들의 오류는 사전확률("유럽에서 전면전은 없었다")에 너무 강하게 매달려, 가능도("침공한다면 이 정도 집결은 당연하다")를 제대로 반영하지 않은 것이다.

- 가능도를 무시한 결론: "집결은 블러핑이다. 침공 확률 여전히 낮다."
- 가능도를 반영한 결론: "이 규모 집결은 침공 가설을 강하게 지지한다. 확률을 올려야 한다."

**가능도의 의미**

가능도 P(D|H)는 "침공 가능성이 얼마냐"를 직접 묻는 게 아니다. **"침공한다는 가설이 맞다면, 지금 내가 본 증거(병력 집결)가 얼마나 자연스럽게 설명되는가"** 를 묻는다.

> 가능도가 높다 = 이 증거는 이 가설이 맞을 때 잘 나타나는 종류의 증거다.

## 3. 의료 진단 예시: 직관이 틀리는 순간

베이즈 정리의 위력을 보여주는 고전 예시다.

**상황**

- 어떤 병의 유병률(전체 인구의 환자 비율) = **1%**
- 검사의 민감도(재현율)(환자를 양성으로 맞힐 확률) = **99%**
- 검사의 특이도(건강한 사람을 음성으로 맞힐 확률) = **95%**

**질문**: 어떤 사람이 이 검사에서 양성으로 나왔다. 이 사람이 실제로 환자일 확률은?

---

### 베이지안으로 먼저 보기

- **H (가설)**: 이 사람이 실제 환자다
- **D (증거)**: 검사 결과가 양성이다

| 개념 | 질문 | 이 사례 |
|---|---|---|
| 사전확률 P(H) | 검사 결과를 보기 전, 환자일 가능성은? | 유병률 = **1%** |
| **가능도 P(D\|H)** | **환자라면, 양성이 나오는 게 자연스러운가?** | **민감도(재현율) = 99% → 매우 높음** |
| 가능도 P(D\|¬H) | 건강하다면, 양성이 나오는 게 자연스러운가? | 1 − 특이도 = 5% → 낮지만 0은 아님 |
| 사후확률 P(H\|D) | 양성을 본 뒤, 실제 환자일 확률은? | **직관보다 훨씬 낮다 — 아래에서 계산** |

직관적으로는 "민감도(재현율)가 99%니까 양성이면 거의 환자"라고 생각하기 쉽다. 그러나 **사전확률(유병률 1%)이 매우 낮다**는 점을 반영하면 결론이 완전히 달라진다.

---

### 혼동행렬로 계산하기

확률로 바로 계산하기 어려울 때는 **10,000명이 검사를 받았다고 상상**하면 쉽다.

1. 유병률 1% → 실제 환자 **100명**, 건강한 사람 **9,900명**
2. 민감도(재현율) 99% → 환자 100명 중 **99명이 양성** (진양성, TP)
3. 특이도 95% → 건강한 9,900명 중 5%인 **495명이 양성** (위양성, FP)
4. 양성 판정 전체: 99 + 495 = **594명**
5. 이 중 실제 환자: `99 / 594 ≈ **17%`**

| | 검사 양성 | 검사 음성 | 합계 |
|---|---:|---:|---:|
| **실제 환자** | TP = 99 | FN = 1 | 100 |
| **실제 건강** | FP = 495 | TN = 9,405 | 9,900 |
| 합계 | 594 | 9,406 | 10,000 |

**사후확률 = TP / (TP + FP) = 99 / 594 ≈ 17%**

민감도(재현율)가 99%인데도 사후확률은 17%다. 왜인가? 건강한 사람이 워낙 많아서, 5%의 위양성률만으로도 거짓 양성(495명)이 진짜 양성(99명)을 압도하기 때문이다.

각 셀의 이름을 정리하면:

| 셀 | 이름 | 의미 |
|---|---|---|
| TP (진양성) | True Positive | 환자를 환자로 올바르게 판정 |
| FN (위음성) | False Negative | 환자를 건강으로 잘못 판정 → **2종 오류** |
| FP (위양성) | False Positive | 건강을 환자로 잘못 판정 → **1종 오류** |
| TN (진음성) | True Negative | 건강을 건강으로 올바르게 판정 |

**혼동행렬에서 읽을 수 있는 지표**

| 지표 | 공식 | 의미 | 값 |
|---|---|---|---:|
| 정확도 Accuracy | `(TP + TN) / 전체` | 전체 검사 중 맞힌 비율 | 95.0% |
| **정밀도 Precision** | `TP / (TP + FP)` | **양성 중 진짜 환자 비율 = 사후확률** | **16.7%** |
| 민감도(재현율) Recall | `TP / (TP + FN)` | 실제 환자 중 잡아낸 비율 = **가능도** | 99.0% |
| 특이도 Specificity | `TN / (TN + FP)` | 건강한 사람 중 음성으로 잘 판정한 비율 | 95.0% |

핵심 혼동 지점: **민감도(재현율)(99%)과 정밀도(16.7%)는 방향이 반대다.**

- 민감도(재현율) = P(양성 | 실제 환자) — 환자의 눈으로 검사를 본다
- 정밀도 = P(실제 환자 | 양성) — 양성 결과의 눈으로 환자를 본다 → **사후확률**

민감도(재현율)가 아무리 높아도, 사전확률(유병률)이 낮으면 정밀도(사후확률)는 낮을 수 있다.

기획자에게 주는 교훈:

> **드문 사건일수록 사전확률을 무시하면 의사결정이 틀린다.** 정책 효과 검증, 사기 탐지, AI 오류 탐지 모두 같다.

## 4. 베이지안 업데이트: 정책 시범사업 사례

사례: **신규 청년 일자리 정책 시범 사업**

기획자가 자주 마주치는 질문은 "이 정책이 성공할 것인가?"가 아니다. 더 정확한 질문은 다음과 같다.

> 지금까지의 정보만으로 이 정책을 확대해도 되는가?

예를 들어 청년 일자리 정책을 10개 지역에서 먼저 시범 운영했다고 하자. 정책의 목적은 단순히 프로그램을 운영하는 것이 아니라, 참여 청년의 취업률을 높이고, 중도탈락을 줄이며, 예산 대비 효과를 만드는 것이다. 따라서 베이지안 업데이트를 하기 전에 먼저 **성공의 기준**을 정해야 한다.

**상황**

- 정책: 신규 청년 일자리 정책 시범 사업
- 결정: 시범 사업을 더 큰 규모로 확대할 것인가?
- 성공 기준: 지역별 시범 사업이 사전에 정한 성과 기준을 달성하면 성공으로 본다.
- 사전 믿음: "성공 확률은 30% 정도" — 비슷한 정책의 성과가 좋지 않았고, 근거도 강하지 않다.
- 새 정보: 시범 사업 10건 중 7건이 성공했다.
- 질문: 이 정보를 본 뒤 성공 확률 추정은 어떻게 바뀌는가?

여기서 중요한 점은 사전 믿음이 단순한 감이 아니라는 것이다. 베이지안 분석에서는 사전 믿음을 **가상의 과거 경험**처럼 표현한다. 이때 0~1 사이의 성공 확률을 나타내기 위해 **베타 분포**를 쓴다. 베타 분포는 "성공할 확률 자체가 얼마나 그럴듯한가"를 나타내는 분포다.

**베타 분포를 직관적으로 읽기**

`Beta(α, β)`에서 α는 성공 쪽 근거, β는 실패 쪽 근거라고 생각하면 된다.

| 표현 | 직관적 의미 | 평균 |
|---|---|---:|
| `Beta(3, 7)` | 성공 3건, 실패 7건을 본 것처럼 약하게 믿음 | `3 / (3 + 7) = 0.30` |
| `Beta(30, 70)` | 성공 30건, 실패 70건을 본 것처럼 강하게 믿음 | `30 / (30 + 70) = 0.30` |

두 경우 모두 평균은 30%다. 그러나 강도는 다르다. `Beta(3, 7)`은 총 10건짜리 믿음이므로 새 데이터에 쉽게 흔들린다. 반면 `Beta(30, 70)`은 총 100건짜리 믿음이므로 같은 새 데이터를 보아도 천천히 바뀐다.

이번 사례에서는 근거가 약하다고 했으므로 사전을 `Beta(3, 7)`로 둔다. 이는 "성공 가능성을 30% 정도로 낮게 보지만, 아직 확신은 강하지 않다"는 뜻이다.

**업데이트 규칙**

베타 분포를 쓰면 업데이트는 매우 단순하다.

1. 사전을 `Beta(α, β)`로 적는다. α = 사전 성공, β = 사전 실패.
2. 새 데이터에서 성공 s건, 실패 f건을 센다.
3. 사후는 `Beta(α + s, β + f)`가 된다.

이번 사례에 대입하면 다음과 같다.

| 구분 | 성공 | 실패 | 분포 | 평균 |
|---|---:|---:|---|---:|
| 사전 믿음 | 3 | 7 | `Beta(3, 7)` | `3 / 10 = 0.30` |
| 새 데이터 | 7 | 3 | - | `7 / 10 = 0.70` |
| 사후 믿음 | 10 | 10 | `Beta(10, 10)` | `10 / 20 = 0.50` |

관찰 데이터만 보면 성공률은 70%다. 그러나 사후 평균은 50%다. 왜 70%가 아니라 50%일까? 기존에 "이 정책은 성공하기 어려울 것"이라는 사전 믿음이 있었기 때문이다. 베이지안 업데이트는 새 데이터를 그대로 믿는 방식이 아니라, **기존 믿음과 새 증거를 합쳐 믿음을 갱신하는 방식**이다.

따라서 이 사례의 해석은 다음과 같다.

- 처음에는 성공 가능성을 30%로 낮게 보았다.
- 그런데 10건 중 7건이 성공했으므로 정책에 긍정적인 증거가 들어왔다.
- 그 결과 성공 확률에 대한 믿음은 30%에서 50% 수준으로 올라갔다.
- 하지만 데이터가 아직 10건뿐이므로 "성공이 확실하다"고 말하기는 어렵다.

## 5. 베타 분포: 확률에 대한 확률

베이지안 업데이트를 하려면 "사전 믿음"을 수식으로 표현해야 한다. 이때 쓰는 도구가 **베타 분포**다.

---

**핵심 질문부터**

동전을 던지면 앞면이 나올 확률이 50%라고 한다. 그런데 그 "50%"를 얼마나 확신하는가?

- 동전을 2번 던져 1번 앞면 → "50% 같은데... 잘 모르겠다"
- 동전을 1000번 던져 500번 앞면 → "50% 맞다, 확신한다"

둘 다 50%지만 **확신의 강도가 다르다.** 이 차이를 표현하는 도구가 베타 분포다.

---

**베타 분포 = "성공 확률에 대한 확신 분포"**

일반 확률분포는 "결과"가 어떻게 퍼져 있는지를 나타낸다.
베타 분포는 **"성공 확률 p 자체"가 얼마나 그럴듯한지**를 나타낸다.

```
일반 분포:  P(매출이 1억이다)
베타 분포:  P(성공 확률이 0.3이다)  ← 확률에 대한 확률
```

---

**`Beta(α, β)` 읽는 법**

α = 지금까지 본 성공 수, β = 지금까지 본 실패 수라고 생각하면 된다.

| 분포 | 성공 | 실패 | 평균 | 모양 |
|--|--:|--:|--:|--|
| `Beta(1, 1)` | 0 | 0 | 50% | 완전히 평평 — "아무것도 모른다" |
| `Beta(3, 7)` | 3 | 7 | 30% | 30% 근처지만 넓게 퍼짐 — 약한 믿음 |
| `Beta(30, 70)` | 30 | 70 | 30% | 30% 근처에 뾰족하게 집중 — 강한 믿음 |

```
Beta(1,1)   →  ──────────────  (뭐든 가능)
Beta(3,7)   →    ~~~▲~~~       (30% 근처 같은데 불확실)
Beta(30,70) →       ▲          (30% 거의 확실)
```

평균은 `α / (α + β)`로 계산한다. α+β가 클수록 곡선이 뾰족해진다.

---

**업데이트가 덧셈으로 끝나는 이유**

새 데이터(성공 s건, 실패 f건)를 받으면 α, β에 그냥 더하면 된다.

```
사전 Beta(3, 7)  +  데이터 7성공 3실패  =  사후 Beta(10, 10)
      α=3, β=7       α+=7, β+=3              α=10, β=10
      평균 30%                                평균 50%
```

수식 유도 없이 **덧셈만으로 베이지안 업데이트가 끝난다.**

한 줄 요약:

> 베타 분포는 "성공 확률이 얼마인지에 대한 나의 믿음"을 곡선으로 표현한 것이다. α가 크면 성공 쪽으로 기울고, α+β가 크면 그 믿음이 강하다.

## 6. 사전이 강하면 어떻게 되는가

사전확률은 평균(방향)뿐 아니라 **강도(얼마나 확신하는가)**도 있다. 같은 30%를 믿더라도 근거가 10건짜리인지 100건짜리인지에 따라 새 데이터의 영향이 완전히 달라진다.

**사전의 강도 = 가상의 과거 관측 수**

| | 약한 믿음 | 강한 믿음 |
|--|--|--|
| 표현 | `Beta(3, 7)` | `Beta(30, 70)` |
| 평균 | 30% | 30% |
| 가상 관측 수 | 10건 | 100건 |
| 의미 | "비슷한 사례를 10건 본 것처럼 믿는다" | "비슷한 사례를 100건 본 것처럼 믿는다" |

---

**같은 데이터를 받으면 사후가 달라진다**

새 데이터: 10건 중 7건 성공 (관측 성공률 70%)

| | 사전 | 새 데이터 | 사후 | 사후 평균 |
|--|--|--|--|--:|
| 약한 믿음 | `Beta(3, 7)` | +7성공 +3실패 | `Beta(10, 10)` | **50.0%** |
| 강한 믿음 | `Beta(30, 70)` | +7성공 +3실패 | `Beta(37, 73)` | **33.6%** |

약한 믿음은 새 데이터에 크게 흔들려 30% → 50%로 이동했다. 강한 믿음은 같은 데이터를 보고도 30% → 33.6%에 머물렀다.

---

**투표 비유**

- 약한 믿음: 이미 10표를 "30%"에 던진 상태. 새로 10표가 "70%"로 들어오면 → 합산 20표 중 균형점인 **50%**
- 강한 믿음: 이미 100표를 "30%"에 던진 상태. 새로 10표가 "70%"로 들어와도 100표를 이길 수 없다 → **33.6%**에 머묾

사전이 강할수록 새 데이터가 믿음을 바꾸려면 **더 많은 증거**가 필요하다.

## 7. 데이터가 쌓이면 결국 데이터가 이긴다

데이터가 적을 때는 사전이 결과에 큰 영향을 준다. 데이터가 쌓이면 사전이 어떤 모양이든 **사후는 데이터에 가까워진다**.

이번 표는 진짜 성공률이 0.7이라고 가정하고, 표본 크기를 10에서 1000까지 늘리면서 사후 평균과 90% 신용 구간이 어떻게 변하는지 보여준다.

진짜 성공률이 70%라고 가정하고, 표본을 늘려가면서 사후 평균과 90% 신용 구간이 어떻게 좁아지는지 보자. 사전은 `Beta(3, 7)`로 고정한다.

| 표본 수 | 성공 수 | 사후 평균 | 하한 5% | 상한 95% |
|---:|---:|---:|---:|---:|
| 10 | 7 | 0.500 | 0.272 | 0.728 |
| 50 | 35 | 0.675 | 0.557 | 0.785 |
| 200 | 140 | 0.695 | 0.641 | 0.748 |
| 1000 | 700 | 0.699 | 0.677 | 0.721 |

**해석**: 데이터가 쌓일수록 사후 평균은 실제 성공률(70%)에 가까워지고, 신용 구간이 좁아진다. 표본이 작을 때는 사전 믿음(30%)의 영향이 남아 평균이 낮다. **데이터가 충분히 쌓이면 어떤 사전을 썼든 결국 데이터가 이긴다.**

## 9. 결정으로 연결하기

분석은 여기서 끝낼 수 있다. 기획은 한 단계 더 가야 한다.

**결정 매뉴얼**

1. 사후 평균을 계산한다.
2. 의사결정 임계값(threshold)을 정한다. 정책의 위험 허용 수준에 따라 다르다.
3. 사후 분포에서 임계값을 넘을 확률을 계산한다.
4. 권고를 작성한다: `Go`, `Hold`, `More data`.

위험이 큰 결정일수록 임계값을 높게 잡는다.

| 결정 유형 | 권장 임계값 |
|---------|-----------|
| 작은 시범 확장 | 0.5 |
| 본 사업 결정 | 0.7 |
| 사람 생명·안전 관련 | 0.9 |

사후 평균만으로 결정하면 불확실성을 무시한다. 더 정확한 방법은 "임계값을 넘을 확률"을 보는 것이다. 아래는 사후 `Beta(10, 10)`에서 임계값별로 계산한 결과다.

| 임계값 | 사후 평균 | P(성공률 > 임계값) | 권고 |
|---:|---:|---:|---|
| 0.3 | 0.500 | 0.981 | **Go** |
| 0.5 | 0.500 | 0.500 | **Hold** (데이터 추가 수집) |
| 0.7 | 0.500 | 0.055 | **Stop or redesign** |
| 0.9 | 0.500 | 0.000 | **Stop or redesign** |

**읽는 법**: 임계값 0.5, 즉 "성공 확률이 절반은 넘어야 확대한다"는 기준이라면 → 확률이 정확히 50%이므로 Hold가 적절하다. 위험이 작은 결정(임계값 0.3)이라면 Go다.

> 임계값은 분석가가 정하는 것이 아니라 **의사결정자가 위험 허용 수준에 따라 정하는 것**이다.
