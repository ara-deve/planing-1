# 7주차. 시나리오 플래닝

## 오늘의 목표

이번 시간에는 미래를 정확히 맞히는 법을 배우지 않는다. **여러 가능한 미래에 대비하는 사고방식**을 익힌다.

수업이 끝나면 다음을 할 수 있어야 한다.

1. 예측과 시나리오 플래닝의 차이를 한 문장으로 설명한다.
2. STEEP 동인을 식별하고 불확실성-영향도로 우선순위를 매긴다.
3. 2×2 매트릭스로 4개 시나리오를 만든다.
4. 윈드 터널링으로 강건한 전략을 찾는다.
5. 시나리오마다 조기 경보 지표를 설계한다.

## 1. 예측 vs 시나리오 플래닝

예측은 "가장 그럴듯한 미래 하나"를 맞히려고 한다. 격변기에는 빗나간다.

시나리오 플래닝은 "가능한 미래 3-4개"를 그려 두고, 어느 것이 와도 **무너지지 않을 전략**을 찾는다.

| 구분 | 예측 (Forecasting) | 시나리오 플래닝 |
|------|------------------|----------------|
| 목적 | 가장 가능성 높은 미래 1개 | 가능한 미래 3-4개 |
| 가정 | 추세가 이어진다 | 추세가 끊길 수 있다 |
| 결과물 | 숫자 하나 | 서사 여러 개 |
| 적합 상황 | 안정적 환경 | 격변기·구조 변화 |

**기획자의 질문**

> 어떤 미래가 와도 살아남는 전략은 무엇인가? 어떤 미래에서만 통하는 전략은 무엇인가?

## 2. STEEP 프레임워크와 시나리오 플래닝 프로세스

동인(Driving Forces)은 미래를 결정하는 큰 흐름이다. 빠뜨리지 않으려면 5개 영역으로 나눠 본다.

| STEEP | 영역 | 예시 |
|-------|------|------|
| **S**ocial | 사회·인구 | 고령화, 세대 가치관 |
| **T**echnological | 기술 | 자율주행, AI |
| **E**conomic | 경제 | 금리, 환율, 성장률 |
| **E**cological | 환경·기후 | 탄소세, 기후 재해 |
| **P**olitical | 정치·규제 | 선거, 규제 강화 |

**시나리오 플래닝 4단계**

1. STEEP 동인을 식별하고 불확실성·영향도를 매긴다 → 핵심 불확실성 2개를 고른다.
2. 두 축의 조합으로 시나리오 4개를 만들고 서사를 붙인다.
3. 전략 후보를 시나리오에 통과시켜(윈드 터널링) 강건한 전략을 찾는다.
4. 시나리오마다 조기 경보 지표를 설계한다.

## 3. 사례: 2035년 한국 도시 모빌리티

**배경**

서울시 교통정책과에서 2035년 도시 모빌리티 시나리오 플래닝을 의뢰했다. 전기차, 자율주행, 탄소 규제 등 불확실한 변수가 많아 하나의 미래를 예측하기보다 여러 시나리오에 대비하려 한다.

이 사례에서 AI와 대화하며 시나리오 플래닝을 완성하는 과정을 보여준다.

### 1단계. 동인 분석 + 핵심 불확실성 선정

동인마다 **불확실성**(얼마나 결과를 모르는가)과 **영향도**(결과가 얼마나 큰가)를 점수로 매긴다. 둘 다 높은 동인이 **시나리오 축 후보**다.

```
영향도 높음
   |
   |  결정 요인         ★ 핵심 불확실성
   |  (이미 정해짐)        (시나리오 축)
   |
   +-------------+-------------
   |
   |  무시 가능         관심사
   |                   (배경 변수)
   |
   +-------------------- 불확실성 높음
```

점수의 신뢰성을 확보하려면 **조사와 점수 산정을 분리**해야 한다. AI에게 한 번에 "점수 매겨줘"라고 하면 근거 없이 숫자를 만들 위험이 있다. 먼저 조사를 시키고, 그 조사 결과를 근거로 점수를 매기게 한다.

**1차 질문 — 조사**

> 2035년 한국 도시 모빌리티에 영향을 미치는 요인들을 조사해줘. STEEP(사회·기술·경제·환경·정치) 영역별로 2-3개씩, 총 12개 동인을 뽑아줘. 각 동인마다 현재 상황과 향후 전망을 실제 보고서·통계·뉴스를 검색해서 정리해줘.

**2차 질문 — 점수 산정**

> 위 조사 자료를 근거로, 각 동인의 불확실성(0-10)과 영향도(0-10) 점수를 매겨줘. 점수마다 "왜 이 점수인지"를 조사 내용을 인용해서 설명해줘. score = 불확실성 × 영향도 순으로 정렬하고, 시나리오 축으로 적합한 독립적인 두 동인을 추천해줘.

**1차 답변 — 조사 결과 (요약)**

| 동인 | 범주 | 현황·전망 |
|------|:---:|----------|

| 대중교통 투자 | 정치 | GTX 개통(2024~), 광역급행철도 확대 논의. 재정 여건에 따라 지연 가능. |

**2차 답변 — 점수 산정**

| 1 | 탄소 규제 강화 | 정치 | 8 | 9 | 72 | 도입 자체는 유력하나 시기·강도가 미정. 시행 시 산업 전반 재편. |
| 2 | 자율주행 기술 성숙도 | 기술 | 8 | 8 | 64 | 성숙 시점 편차 7년. 실현 시 택시·물류·도시설계 변화. |
| 3 | 충전 인프라 확충 속도 | 기술 | 7 | 9 | 63 | 목표 대비 보급률 50%. EV 전환의 병목. |
| 4 | 리튬 공급 리스크 | 경제 | 9 | 6 | 54 | 수급 불균형 심각하나 대체 기술이 영향도를 낮춤. |
| 5 | 반(反)EV 정책 반발 | 정치 | 7 | 7 | 49 | 글로벌 선례 있음. 한국 산업구조상 반발 여지 큼. |

**시나리오 축 추천**
- **X축: 탄소 규제 강화** (정치, score 72) — 강화 ↔ 약화
- **Y축: 충전 인프라 확충 속도** (기술, score 63) — 빠름 ↔ 느림

자율주행(score 64)도 후보지만 충전 인프라와 같은 기술 범주여서 상관이 있다. 탄소 규제(정치)와 충전 인프라(기술)는 서로 다른 영역이므로 독립적이다.

**핵심**: 조사 → 점수 산정을 분리하면 **점수마다 추적 가능한 근거**가 붙는다. 최종적으로 "이 근거가 타당한가?"는 사람이 판단해야 한다.

### 2단계. 2×2 시나리오 매트릭스

두 축의 극단을 조합하면 4개의 미래가 나온다. 각 시나리오에 **짧은 이름**과 **서사**(누가 이기고, 누가 지는가)를 붙인다. 확률은 매기지 않는다 — 가능성을 펼쳐 두는 것이 목적이다.

**AI에게 이렇게 질문합니다**

> 탄소 규제(강화/약화)와 충전 인프라(빠름/느림) 두 축으로 시나리오 4개를 만들어줘. 각 시나리오에 한국어 이름(2-3단어)과 핵심 서사(2문장)를 붙여줘. 서사에는 누가 이기고 누가 지는지 적어줘. 2×2 매트릭스 그림도 그려줘.

**AI의 답변**

```
                        충전 인프라 빠름 ↑
                              │
                              │
         ┌────────────────────┼────────────────────┐
         │                    │                    │
         │    기술 경쟁        │    녹색 질주        │
         │   (규제↓ · 인프라↑)  │   (규제↑ · 인프라↑)  │
         │                    │                    │
 탄소규제  ├────────────────────┼────────────────────┤ 탄소규제
  약화 ←  │                    │                    │  → 강화
         │    느린 전환        │    규제 공회전       │
         │   (규제↓ · 인프라↓)  │   (규제↑ · 인프라↓)  │
         │                    │                    │
         └────────────────────┼────────────────────┘
                              │
                              │
                        충전 인프라 느림 ↓
```

| 시나리오 | 규제 | 인프라 | 핵심 서사 |
|---------|:---:|:---:|----------|
| **녹색 질주** | 강화 ↑ | 빠름 ↑ | 강한 탄소세와 빠른 충전망 확충이 EV 전환을 가속한다. EV 제조사·충전 사업자가 이기고, 내연기관 부품사가 진다. |
| **기술 경쟁** | 약화 ↓ | 빠름 ↑ | 규제는 약하지만 민간 기술 경쟁으로 충전망이 깔린다. 빅테크·스타트업이 이기고, 규제에 기댄 공기업이 진다. |
| **규제 공회전** | 강화 ↑ | 느림 ↓ | 규제는 강한데 인프라가 못 따라와 시민 불만이 폭발한다. 정치적 백래시가 생기고, 정부 신뢰가 진다. |
| **느린 전환** | 약화 ↓ | 느림 ↓ | 모든 변화가 미뤄진다. 기존 자동차 산업·정유사가 시간을 벌고, EV 스타트업이 진다. |

**핵심**: 4개 시나리오는 "좋은 미래 vs 나쁜 미래"가 아니다. **이해관계자마다 유리한 시나리오가 다르다**는 점이 중요하다.

### 3단계. 윈드 터널링: 강건한 전략 찾기

윈드 터널링은 풍동(風洞) 실험에서 따온 말이다. 전략 후보를 여러 시나리오라는 "바람"에 통과시켜, 어느 바람에도 무너지지 않는 전략을 찾는다.

- **min**(최저 점수)이 가장 높은 전략 = **가장 강건한 전략**
- mean은 높지만 min이 낮은 전략 = **도박형 전략**

**AI에게 이렇게 질문합니다**

> 아래 전략 3개를 위 4개 시나리오에서 평가해줘. A: 공격적 EV 공장 투자, B: 하이브리드 + 대중교통 투자, C: 관망 + 기술 라이선스. 각 전략의 시나리오별 점수(-2 ~ +2)를 매기고, min과 mean으로 강건성을 비교해줘.

**AI의 답변**

| 전략 | 녹색 질주 | 기술 경쟁 | 규제 공회전 | 느린 전환 | min | mean |
|------|:---:|:---:|:---:|:---:|:---:|:---:|
| A: 공격적 EV 투자 | +2 | -2 | +1 | -1 | **-2** | 0.00 |
| B: 하이브리드+대중교통 | +1 | +1 | +1 | +2 | **+1** | 1.25 |
| C: 관망+라이선스 | -1 | +2 | -1 | +1 | **-1** | 0.25 |

**추천: 전략 B**

- 전략 B는 min=+1로 어느 시나리오에서도 무너지지 않는다. **강건한 전략**.
- 전략 A는 녹색 질주에서 최고점(+2)이지만 기술 경쟁에서 -2. **도박형**.
- 전략 C는 기술 경쟁에서만 +2, 나머지에서 약하다. **편향된 전략**.

**핵심**: 기획자가 할 일은 "가장 점수가 높은 전략"이 아니라 **"최악의 경우에도 견딜 수 있는 전략"**을 고르는 것이다.

### 4단계. 조기 경보 지표

강건한 전략을 골랐지만, 세상은 바뀐다. 어떤 시나리오가 현실로 다가오는지 알려주는 **신호**를 미리 정해두면, 전략을 적시에 조정할 수 있다.

좋은 지표의 조건:
1. **측정 가능**하다 ("EV 등록 비율" ○, "분위기가 바뀐다" ✕)
2. **데이터 출처**가 있다 (어디서 분기마다 볼 수 있는가)
3. **임계값**이 있다 ("분기 등록 비율 30% 초과 시")

**AI에게 이렇게 질문합니다**

> 4개 시나리오(녹색 질주, 기술 경쟁, 규제 공회전, 느린 전환)가 현실이 되어가는지 알려줄 조기 경보 지표를 시나리오마다 2개씩 만들어줘. 각 지표에 데이터 출처와 임계값을 넣어줘.

**AI의 답변**

| 시나리오 | 지표 | 데이터 출처 | 임계값 |
|---------|------|-----------|--------|
| 녹색 질주 | 신규 EV 등록 비율 | 국토교통부 자동차등록 통계 | 분기 30% 초과 |
| 녹색 질주 | 공공 충전소 증가율 | 한국전력 충전인프라 현황 | 연 25% 초과 |
| 기술 경쟁 | 민간 충전 투자액 | 전자공시시스템(DART) | 연 1조원 초과 |
| 기술 경쟁 | 자율주행 시범 도시 수 | 국토부 자율주행 현황 | 5개 도시 이상 |
| 규제 공회전 | 충전소 증가율 | 한국전력 충전인프라 현황 | 연 5% 미만 |
| 규제 공회전 | 교통 민원 건수 | 국민신문고 | 전년 대비 20% 증가 |
| 느린 전환 | 탄소세 법안 진행 | 국회 의안정보시스템 | 2027년까지 미통과 |
| 느린 전환 | 내연기관 신차 판매량 | 한국자동차산업협회(KAMA) | 전년 대비 횡보·증가 |

**핵심**: 조기 경보 지표는 "미래를 맞히는 도구"가 아니라 **"어떤 시나리오로 가고 있는지 빨리 감지하는 레이더"**다. 분기마다 점검하면 전략 조정 타이밍을 놓치지 않는다.

## 4. 실습: 나만의 시나리오 플래닝

위 사례에서 본 4단계를 **자기 주제**로 직접 진행한다. AI에게 질문하고, 받은 답변을 결과 셀에 붙여넣는다.

**예시 주제**

- 2035년 한국 청년 일자리
- 2030년 지역 관광 산업
- 2030년 대학 교육 모델
- 2030년 고령자 돌봄 서비스

### 실습 1단계. 동인 분석 + 핵심 불확실성 선정

**1차 질문 — 조사**

> 나는 [주제, 예: 2035년 한국 청년 일자리]에 대한 시나리오 플래닝을 하려고 해. STEEP(사회·기술·경제·환경·정치) 영역별로 2-3개씩, 총 12개 동인을 뽑아줘. 각 동인마다 현재 상황과 향후 전망을 실제 보고서·통계·뉴스를 검색해서 정리해줘.

**2차 질문 — 점수 산정**

> 위 조사 자료를 근거로, 각 동인의 불확실성(0-10)과 영향도(0-10) 점수를 매겨줘. 점수마다 "왜 이 점수인지"를 조사 내용을 인용해서 설명해줘. score = 불확실성 × 영향도 순으로 정렬하고, 시나리오 축으로 적합한 독립적인 두 동인을 추천해줘.

① 안정 번영 (공급망 강 / 기후 느림)

- 서사

- 서사

③ 확보 경쟁 (공급망 약 / 기후 느림)

- 서사
기후는 안정적이지만 보호무역으로 공급망이 분절되면서 국가 간 식량 확보 경쟁이 심화된다.
외교력이 강한 국가는 유리해지고, 한국처럼 수입 의존도가 높은 국가는 협상에서 불리해진다.

④ 복합 위기 (공급망 약 / 기후 빠름)

- 서사
기후위기로 생산이 줄어드는 동시에 공급망까지 붕괴되어 식량 확보 자체가 어려워진다.
식량 수출국과 자급률이 높은 국가는 생존하고, 한국과 같은 수입 의존 국가는 심각한 위기에 직면한다.



### 실습 3단계. 윈드 터널링

**AI에게 이렇게 질문하세요**

> 내 전략 후보 3개를 줄게: [A], [B], [C]. 각 전략을 위 4개 시나리오에서 -2~+2로 점수 매기고, min과 mean으로 어떤 전략이 가장 강건한지 비교해줘.

**결과**

1️. 평가 기준 정리
+2: 매우 효과적
+1: 효과 있음
0: 영향 제한적
-1: 비효율/부담 발생
-2: 실패 가능성 높음
2️. 시나리오별 전략 점수
시나리오 재정리
안정 번영 (공급망 강 / 기후 느림)
비용 압박 (공급망 강 / 기후 빠름)
확보 경쟁 (공급망 약 / 기후 느림)
복합 위기 (공급망 약 / 기후 빠름)
A : 공급망 다변화 전략
안정 번영: +1
- 이미 안정적이라 추가 효과는 제한적
비용 압박: +2
- 다양한 공급처 확보 → 가격 충격 완화
확보 경쟁: +2
- 공급망 분절 상황에서 가장 강력한 대응
복합 위기: +1
- 일부 확보 가능하지만 한계 존재

A 총평: 외부 리스크 대응에 매우 강함

B : 국내 생산 강화 전략
안정 번영: -1
- 비용 높고 비효율 (수입이 더 유리)
비용 압박: +1
- 가격 상승 시 일부 완충 역할
확보 경쟁: +2
- 수입 어려울 때 핵심 역할
복합 위기: +2
- 최악 상황에서 생존 전략

B 총평: 위기 대응에 강하지만 평시 비효율

C : 수요 관리 전략
안정 번영: +1
- 장기적으로 효율 개선
비용 압박: +2
- 소비 감소 → 가격 부담 완화
확보 경쟁: +1
- 수요 감소로 일부 대응 가능
복합 위기: +2
- 부족 상황에서 가장 현실적 대응

C 총평: 모든 상황에서 안정적 효과

3️. 종합 점수 (min / mean)
전략	점수 리스트	min	mean
A (공급망)	1, 2, 2, 1	1	1.5
B (국내 생산)	-1, 1, 2, 2	-1	1.0
C (수요 관리)	1, 2, 1, 2	1	1.5

4️. 강건성(robustness) 평가
최종 결과

- 가장 강건한 전략: A, C (공동 1위)

- 해석
A : 공급망 전략
평균 높음 (1.5)
최악 상황에서도 +1 유지
- “절대 망하지 않는 전략”
C : 수요 관리 전략
모든 시나리오에서 플러스
- “가장 안정적인 전략”
B : 국내 생산 전략
위기에서는 강함
평시에는 비효율 (-1 존재)
- “보험형 전략 (위기 대비용)”

5️. 결론 

분석 결과, 공급망 다변화 전략(A)과 수요 관리 전략(C)은 모든 시나리오에서 안정적인 성과를 보이며 가장 강건한 전략으로 나타났다. 반면 국내 생산 강화 전략(B)은 위기 상황에서는 높은 효과를 보이지만 평시에는 비효율성이 존재하여 보완적 전략으로 활용하는 것이 적절하다.


### 실습 4단계. 조기 경보 지표

**AI에게 이렇게 질문하세요**

> 4개 시나리오가 현실이 되어가는지 알려줄 조기 경보 지표를 시나리오마다 2개씩 만들어줘. 각 지표에 데이터 출처와 임계값을 넣어줘.

**결과**

1️. 안정 번영 (공급망 강 / 기후 느림)
지표 1: 글로벌 곡물 가격 안정성
설명: 식량 시장이 안정적으로 유지되는지 확인
데이터 출처: FAO 식량가격지수 (FAO Food Price Index)
임계값:
기준: 최근 5년 평균 ±10% 이내 유지 / 20% 이상 상승 시 → 다른 시나리오 전환 가능
지표 2: 주요 곡물 재고율 (Stock-to-Use Ratio)
설명: 글로벌 공급 안정성 지표
데이터 출처: USDA WASDE 보고서
임계값:
안정 기준: 25% 이상 유지 / 20% 이하 하락 시 → 공급 불안 신호

2️. 비용 압박 (공급망 강 / 기후 빠름)
지표 1: 곡물 가격 상승률
설명: 기후 영향으로 인한 가격 상승 감지
데이터 출처: FAO
임계값:연간 15% 이상 상승 시 → 비용 압박 시나리오 진입
지표 2: 주요 생산국 기후 이상 지수
설명: 생산 감소 위험 탐지
데이터 출처: NOAA 기후 데이터
임계값:주요 곡창지대 이상기후 발생률 30% 이상 증가

3️. 확보 경쟁 (공급망 약 / 기후 느림)
지표 1: 식량 수출 제한 조치 수
설명: 보호무역 확대 여부
데이터 출처: WTO 무역 통계
임계값:연간 10건 이상 수출 제한 발생
지표 2: 한국의 식량 수입선 집중도 (HHI)
설명: 특정 국가 의존도 측정
데이터 출처: KOSIS
임계값:HHI 0.25 이상 → 고집중 구조 (위험)

4️. 복합 위기 (공급망 약 / 기후 빠름)
지표 1: 글로벌 곡물 가격 급등 + 재고 감소 동시 발생
설명: 위기 복합 발생 여부
데이터 출처: FAO + USDA
임계값:가격 30%↑ + 재고율 20%↓ 동시 발생
지표 2: 한국 식량 수입 물량 감소율
설명: 실제 위기 체감 지표
데이터 출처: 관세청
임계값:전년 대비 15% 이상 감소 시 → 위기 진입

2️. 핵심 정리 

본 연구에서는 시나리오의 현실화 가능성을 조기에 감지하기 위해 가격, 재고, 기후, 무역 구조 등 다양한 지표를 활용한 조기 경보 체계를 제시하였다. 특히 단일 지표가 아닌 복합 지표(가격 상승과 재고 감소 동시 발생 등)를 활용함으로써 보다 정확한 위기 판단이 가능하다.
