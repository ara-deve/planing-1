# 10주차. 실행 계획과 리스크 관리

## 오늘의 목표

이번 시간에는 프로젝트 관리 이론을 길게 외우지 않는다. 대신 좋은 전략이 실행 단계에서 무너지지 않도록 **목표, 과제, 책임, 리스크, 조기 경보 지표**를 AI와 함께 문서화하는 방법을 익힌다.

수업이 끝나면 다음을 할 수 있어야 한다.

1. 전략-실행 갭이 왜 생기는지 설명한다.
2. OKR로 목표와 핵심 결과를 쓴다.
3. WBS로 실행 과제를 쪼갠다.
4. RACI로 책임을 분명히 한다.
5. 리스크 레지스터와 대응 전략을 작성한다.
6. 조기 경보 지표로 실행 중 점검 기준을 만든다.


## 1. 좋은 전략이 왜 실패하는가

좋은 전략 문서가 있어도 실행은 자주 실패한다. 이유는 전략이 틀려서가 아니라, 실행 조건이 문서화되지 않았기 때문이다.

| 실패 패턴 | 증상 | 필요한 도구 |
|---|---|---|
| 목표가 추상적임 | 모두 동의하지만 아무도 움직이지 않음 | OKR |
| 일이 크고 모호함 | 어디서 시작할지 모름 | WBS |
| 책임이 흐림 | 회의는 많지만 결정이 늦음 | RACI |
| 리스크가 늦게 보임 | 사고가 난 뒤 대응함 | 리스크 레지스터 |
| 신호가 없음 | 계획을 바꿀 타이밍을 놓침 | 조기 경보 지표 |

10주차의 핵심 질문은 이것이다.

> 이 전략을 다음 8주 동안 누가, 무엇을, 어떤 위험을 관리하며 실행할 것인가?


## 2. 실행 계획 5단계

이번 장에서는 하나의 사례를 5단계로 완성한다. 아래 용어들은 프로젝트 관리에서 자주 쓰는 영어 약어이므로, 원어와 뜻을 같이 기억한다.

| 단계 | 영어 원문 | 한국어 뜻 | 이 단계에서 정하는 것 | AI의 역할 |
|---:|---|---|---|---|
| 1 | OKR: Objectives and Key Results | 목표와 핵심 결과 | 무엇을 성공으로 볼지 숫자로 정한다 | 목표 문장과 측정 가능한 KR 제안 |
| 2 | WBS: Work Breakdown Structure | 작업 분해 구조 | 큰 일을 관리 가능한 작은 작업으로 쪼갠다 | Level 2, Level 3 작업 목록 제안 |
| 3 | RACI: Responsible, Accountable, Consulted, Informed | 역할·책임 배분표 | 각 작업을 누가 실행하고 누가 승인할지 정한다 | 작업별 책임자와 승인자 정리 |
| 4 | Risk Register | 리스크 목록표 | 발생 가능한 위험, 확률, 영향, 대응안을 정한다 | 위험 식별, 점수화, 대응안 제안 |
| 5 | Early Warning Indicators | 조기 경보 지표 | 문제가 커지기 전에 볼 신호와 기준값을 정한다 | 점검 지표, 확인 주기, 임계값 설계 |

5단계는 서로 따로 노는 문서가 아니다. **OKR이 목표를 정하고, WBS가 일을 쪼개며, RACI가 책임을 붙이고, 리스크 레지스터가 위험을 관리하고, 조기 경보 지표가 계획 수정 시점을 알려준다.**

코드는 필수가 아니다. 필요한 경우에만 AI에게 표 정리나 매트릭스 시각화를 요청한다.

## 3. 사례: 지역 청년 일자리 매칭 플랫폼 실행 계획

**배경**

한 지자체가 지역 중소기업과 청년 구직자를 연결하는 일자리 매칭 플랫폼을 8주 안에 시범 출시하려 한다. 전략 방향은 이미 정해졌지만, 실행 계획은 아직 모호하다.

**전략 문장**

> 지역 청년이 첫 직장을 더 빠르게 찾도록, 기업 수요와 청년 역량을 데이터 기반으로 연결한다.

**실행 질문**

> 8주 안에 시범 서비스를 출시하려면 어떤 목표, 작업, 책임, 리스크 관리 체계가 필요한가?


### 1단계. OKR 작성

OKR은 실행팀이 같은 방향을 보게 만드는 문장이다.

- Objective: 기억하기 쉬운 정성 목표
- Key Results: 달성 여부를 판단할 수 있는 3~5개 결과 지표

**AI에게 이렇게 질문하세요**

```text
아래 전략을 8주 시범사업 OKR로 바꿔줘.
전략: 지역 청년이 첫 직장을 더 빠르게 찾도록, 기업 수요와 청년 역량을 데이터 기반으로 연결한다.
조건:
- Objective는 한 문장, 정성적이고 기억하기 쉽게
- Key Results는 4개
- 각 KR은 숫자로 측정 가능하게
- 너무 쉬운 목표가 아니라 70% 달성해도 의미 있는 수준으로
```


**AI의 답변 예시**

| 구분 | 내용 |
|---|---|
| Objective | 지역 청년과 중소기업이 8주 안에 실제 채용 대화를 시작하게 만든다 |
| KR1 | 참여 기업 40곳을 확보한다 |
| KR2 | 청년 구직자 300명을 등록시킨다 |
| KR3 | 기업-청년 매칭 120건을 만든다 |
| KR4 | 매칭 후 면접 전환율 35%를 달성한다 |

**점검 질문**

- KR이 활동량이 아니라 결과를 측정하는가?
- 8주 안에 측정 가능한가?
- 담당자가 숫자를 보고 행동을 바꿀 수 있는가?


### 2단계. WBS로 일 쪼개기

WBS는 **Work Breakdown Structure**의 약자다. 직역하면 "작업 분해 구조"이고, 큰 일을 실행 가능한 작은 작업으로 나누는 표를 뜻한다. 여기서 중요한 것은 일정을 먼저 쓰는 것이 아니라, **해야 할 일의 덩어리를 빠짐없이 쪼개는 것**이다.

| 구분 | 의미 | 예시 |
|---|---|---|
| Level 1 | 전체 프로젝트 | 청년 일자리 매칭 플랫폼 8주 시범사업 |
| Level 2 | 큰 작업 묶음 | 기업 모집, 청년 모집, 플랫폼 준비, 매칭 운영, 성과 평가 |
| Level 3 | 실제 관리할 세부 작업 | 대상 기업 리스트 작성, 제안서 발송, 신청 폼 설계, 1차 매칭 실행 |

WBS와 일정표는 다르다. WBS는 **무슨 일을 해야 하는가**를 쪼개는 문서이고, 일정표는 그 일을 **언제 할 것인가**를 배치하는 문서다. 수업에서는 8주 계획이므로 Level 2와 Level 3까지만 작성하고, 각 Level 3 작업에 기간과 산출물을 붙인다.

**AI에게 이렇게 질문하세요**

```text
위 OKR을 달성하기 위한 8주 WBS를 만들어줘.
조건:
- Level 2는 5개 안팎으로
- 각 Level 2마다 Level 3 작업 3~4개
- 각 Level 3 작업에 담당 역할, 기간(주차), 산출물을 넣어줘
- 표 형식으로 정리해줘
```


**AI의 답변 예시**

| Level 2 | Level 3 작업 | 담당 역할 | 기간 | 산출물 |
|---|---|---|---|---|
| 기업 모집 | 대상 기업 리스트 작성 | 기업협력 담당 | 1주 | 기업 100곳 리스트 |
| 기업 모집 | 참여 제안서 발송 | 기업협력 담당 | 1~2주 | 발송 기록 |
| 청년 모집 | 대학·기관 홍보 채널 확보 | 홍보 담당 | 1~2주 | 홍보 채널 목록 |
| 플랫폼 준비 | 신청 폼과 매칭 기준 설계 | 운영 담당 | 2~3주 | 신청 폼, 매칭 기준표 |
| 매칭 운영 | 1차 매칭 실행 | 운영 담당 | 5주 | 매칭 후보표 |
| 성과 평가 | 면접 전환율 집계 | 데이터 담당 | 7~8주 | 성과 대시보드 |

WBS는 세부 작업을 끝없이 늘리는 문서가 아니다. **이번 기간에 실제로 관리할 작업만** 남긴다.


### 3단계. RACI로 책임 분리

RACI는 작업마다 누가 실행하고, 누가 승인하고, 누구에게 의견을 듣고, 누구에게 알릴지 정한다.

| 글자 | 의미 | 원칙 |
|---|---|---|
| R | Responsible: 직접 수행 | 1명 이상 가능 |
| A | Accountable: 최종 책임·승인 | 반드시 1명 |
| C | Consulted: 사전 협의 | 필요한 만큼 |
| I | Informed: 결과 공유 | 필요한 만큼 |

**AI에게 이렇게 질문하세요**

```text
위 WBS의 핵심 작업 6개를 골라 RACI 매트릭스를 만들어줘.
역할은 사업총괄, 기업협력 담당, 청년모집 담당, 플랫폼 담당, 데이터 담당, 외부 파트너로 해줘.
조건:
- 각 작업마다 A는 반드시 1명만
- R과 A가 불분명한 작업은 수정 이유를 적어줘
```


**AI의 답변 예시**

| 핵심 작업 | 사업총괄 | 기업협력 | 청년모집 | 플랫폼 | 데이터 | 외부 파트너 |
|---|:---:|:---:|:---:|:---:|:---:|:---:|
| 참여 기업 확보 | A | R | I | I | I | C |
| 청년 구직자 모집 | A | I | R | I | I | C |
| 신청 폼 구축 | A | C | C | R | C | I |
| 매칭 기준 확정 | A | C | C | R | R | C |
| 1차 매칭 실행 | A | R | R | R | C | I |
| 성과 집계 | A | I | I | C | R | I |

위 표의 열은 사람 이름이 아니라 **역할**이다. 실제 사업에서는 역할마다 담당자 이름을 붙여 읽으면 된다.

| 역할 | 실제 담당자 예시 |
|---|---|
| 사업총괄 | 김민수 팀장 |
| 기업협력 | 이지은 매니저 |
| 청년모집 | 최서연 매니저 |
| 플랫폼 | 박준호 PM |
| 데이터 | 정하늘 분석가 |
| 외부 파트너 | 지역상공회의소 담당자 |

예를 들어 **참여 기업 확보** 행은 이렇게 해석한다.

| 담당자 | RACI | 실제 행동 |
|---|:---:|---|
| 김민수 팀장 | A | 참여 기업 목표 수와 제안 조건을 최종 승인한다. 목표 미달이면 방향을 바꿀 책임도 진다. |
| 이지은 매니저 | R | 기업 후보 리스트를 만들고, 연락하고, 제안서를 보내고, 참여 의사를 확인한다. |
| 최서연 매니저 | I | 어떤 기업이 참여하는지 결과를 공유받는다. 청년모집 메시지를 맞추기 위해 필요하다. |
| 박준호 PM | I | 기업 신청 현황을 공유받는다. 플랫폼 준비 일정에 반영하기 위해 필요하다. |
| 정하늘 분석가 | I | 확보된 기업 수와 직무 정보를 공유받는다. 이후 성과 집계에 사용한다. |
| 지역상공회의소 담당자 | C | 기업 접촉 채널과 추천 기업에 대해 의견을 준다. 직접 최종 책임을 지지는 않는다. |

따라서 참여 기업 확보는 한 사람이 혼자 하는 일이 아니다. 여러 사람이 관여하지만, **직접 뛰는 사람은 R, 최종 책임자는 A, 의견을 주는 사람은 C, 결과를 공유받는 사람은 I**로 구분한다.

**핵심 규칙**: A가 두 명이면 결정이 밀린다. 한 작업의 최종 승인자는 반드시 한 명이다.


### 4단계. 리스크 레지스터 작성

리스크는 막연한 걱정이 아니라 **발생 가능성과 영향도를 가진 사건**이다.

| 컬럼 | 의미 |
|---|---|
| ID | R1, R2... |
| 리스크 | 일어날 수 있는 사건 |
| 분류 | 일정, 비용, 기술, 자원, 외부, 조직 |
| 확률 | 0~1 |
| 영향 | 1~5 |
| 점수 | 확률 × 영향 |
| 대응 전략 | 회피, 완화, 전가, 수용 |
| 대응안 | 실제 행동 |
| 트리거 | 조기 경보 신호 |

확률과 영향은 AI가 임의로 정답을 맞히는 숫자가 아니다. **AI는 초안을 만들고, 사람은 근거를 확인해 조정한다.** 수업에서는 아래 기준으로 간단히 점수화한다.

**확률 점수 기준**

| 확률 | 의미 | 판단 기준 예시 |
|---:|---|---|
| 0.1 | 매우 낮음 | 비슷한 사업에서 거의 발생하지 않았고, 현재 징후도 없음 |
| 0.3 | 낮음 | 발생 가능성은 있지만 관리 경험이나 대안이 있음 |
| 0.5 | 보통 | 발생할 수도 있고 안 할 수도 있음. 근거가 엇갈림 |
| 0.7 | 높음 | 이미 지연, 반대, 인력 부족 같은 징후가 보임 |
| 0.9 | 매우 높음 | 특별한 조치가 없으면 거의 발생할 것으로 보임 |

**영향 점수 기준**

| 영향 | 의미 | 판단 기준 예시 |
|---:|---|---|
| 1 | 작음 | 일부 작업만 조정하면 됨. 전체 목표에는 거의 영향 없음 |
| 2 | 낮음 | 일정이나 품질에 작은 영향이 있음 |
| 3 | 보통 | 핵심 KR 일부가 흔들림. 추가 인력이나 일정 조정이 필요함 |
| 4 | 큼 | 출시, 모집, 예산, 신뢰도 중 하나에 큰 차질이 생김 |
| 5 | 치명적 | 사업 중단, 법적 문제, 핵심 목표 실패로 이어질 수 있음 |

예를 들어 참여 기업 확보 부족은 이미 유사 사업에서 자주 나타났고 현재 기업 반응도 약하다면 확률을 0.6~0.7로 둔다. 이 문제가 생기면 매칭 자체가 어려워지므로 영향은 5점으로 둘 수 있다.

**AI에게 이렇게 질문하세요**

```text
위 청년 일자리 매칭 플랫폼 8주 시범사업의 주요 리스크 10개를 식별해줘.
각 리스크에 분류, 확률(0~1), 영향(1~5), 점수(확률×영향), 대응 전략(회피/완화/전가/수용), 대응안을 넣어줘.
확률과 영향은 위 기준표에 따라 추정하고, 각 점수의 근거를 한 문장으로 설명해줘.
점수가 높은 순서로 정렬해줘.
```


**AI의 답변 예시**

| ID | 리스크 | 분류 | 확률 | 영향 | 점수 | 점수 근거 | 전략 | 대응안 |
|---|---|---|---:|---:|---:|---|---|---|
| R1 | 참여 기업 확보가 목표보다 부족함 | 외부 | 0.6 | 5 | 3.0 | 기업 반응이 약하면 매칭 자체가 어려워짐 | 완화 | 기존 협회 채널과 추천 인센티브 병행 |
| R2 | 청년 신청자는 많지만 적합 직무가 부족함 | 외부 | 0.5 | 4 | 2.0 | 신청 수요와 기업 직무가 어긋나면 면접 전환율이 하락함 | 완화 | 직무군을 사전 분류하고 부족 직무 기업 우선 모집 |
| R3 | 개인정보 처리 기준 미비로 출시 지연 | 기술/법무 | 0.3 | 5 | 1.5 | 가능성은 낮아도 발생하면 서비스 공개가 막힐 수 있음 | 회피 | 신청 폼 공개 전 법무 검토 완료 |
| R4 | 매칭 기준이 불투명하다는 민원 발생 | 조직 | 0.4 | 4 | 1.6 | 기준 설명이 부족하면 신뢰도와 운영 일정에 타격이 있음 | 완화 | 기준표 공개, 이의제기 절차 마련 |
| R5 | 플랫폼 개발 일정 지연 | 일정 | 0.4 | 4 | 1.6 | 개발 지연은 모집과 매칭 일정 전체를 밀리게 할 수 있음 | 완화 | 노코드 폼으로 MVP 우선 출시 |

리스크 레지스터는 겁을 주는 문서가 아니다. **무엇을 먼저 관리해야 하는지 정하는 우선순위 표**다.


### 5단계. 대응 전략과 조기 경보 지표

리스크 대응은 네 가지로 나눈다.

| 전략 | 의미 | 예시 |
|---|---|---|
| 회피 | 리스크 원인 제거 | 법적 위험 기능은 출시 범위에서 제외 |
| 완화 | 확률이나 영향 감소 | 백업 채널 확보 |
| 전가 | 제3자에게 이전 | 보험, 외주 계약 조항 |
| 수용 | 감수하고 모니터링 | 영향이 작은 일정 변동 |

**AI에게 이렇게 질문하세요**

```text
위 리스크 중 점수 상위 5개에 대해 조기 경보 지표를 만들어줘.
각 지표에는 측정 방법, 확인 주기, 임계값, 임계값을 넘었을 때 실행할 대응 행동을 넣어줘.
표 형식으로 정리해줘.
```


**AI의 답변 예시**

| 리스크 | 조기 경보 지표 | 주기 | 임계값 | 대응 행동 |
|---|---|---|---|---|
| 기업 확보 부족 | 확정 참여 기업 수 | 주간 | 2주차 15곳 미만 | 협회·상공회의소 채널 즉시 투입 |
| 직무 불일치 | 청년 희망 직무 대비 기업 직무 비율 | 주간 | 특정 직무 수요/공급 2배 이상 | 부족 직무 기업 집중 모집 |
| 개인정보 이슈 | 법무 검토 미완료 항목 수 | 주간 | 3주차까지 미완료 1건 이상 | 해당 기능 출시 범위 제외 |
| 플랫폼 지연 | 핵심 폼 테스트 완료율 | 주간 | 4주차 80% 미만 | 노코드 MVP로 전환 |
| 매칭 민원 | 기준 설명 요청 건수 | 주간 | 주 10건 초과 | 기준표와 FAQ 공개 |

조기 경보 지표는 KPI와 다르다. KPI가 성과를 보는 지표라면, 조기 경보 지표는 **문제가 커지기 전에 계획을 바꾸게 하는 신호**다.

## 4. 실습: 나만의 실행 계획

관심 있는 정책 또는 사업 하나를 고른다. 아래 5단계를 자기 주제로 반복한다.

**예시 주제**

- 지역 관광 축제 실행 계획
- 공공 돌봄 앱 시범사업
- 대학 비교과 프로그램 개편
- 소상공인 디지털 전환 지원사업
- 탄소 감축 설비 보급사업

실습의 목표는 멋진 전략 문장이 아니라 **다음 8주 동안 실행 가능한 관리 문서**를 만드는 것이다.
